```python
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import NearMiss

###########################################################

def Random_Under_Sampler(akut):
    
    x = akut.drop(columns=['MRA_previous'])
    y = akut['MRA_previous']
    
    rus = RandomUnderSampler()
    x_rus, y_rus = rus.fit_resample(x, y)
    akut_rus = pd.concat([x_rus, y_rus], axis=1)
    
    print("RUS")
    print('original akutset shape:', Counter(y))
    print('Resample akutset shape', Counter(y_rus))
    
    return akut_rus

###########################################################

def Random_Over_Sampler(akut):
    
    x = akut.drop(columns=['MRA_previous'])
    y = akut['MRA_previous']
    
    ros = RandomOverSampler(random_state=42)
    x_ros, y_ros = ros.fit_resample(x, y)
    akut_ros = pd.concat([x_ros, y_ros], axis=1)
    
    print("ROS")
    print('Original akutset shape', Counter(y))
    print('Resample akutset shape', Counter(y_ros))
    
    return akut_ros

###########################################################

def TomekLinks_Under_Sampler(akut):
    
    x = akut.drop(columns=['MRA_previous'])
    y = akut['MRA_previous']
    
    tl = TomekLinks(sampling_strategy='majority')
    x_tl, y_tl = tl.fit_resample(x, y)
    akut_tl = pd.concat([x_tl, y_tl], axis=1)

    print("TL")
    print('Original akutset shape:', Counter(y))
    print('Resample akutset shape:', Counter(y_tl))
    
    return akut_tl

###########################################################

def SyntheticMinority_Over_Sampler(akut):
    
    x = akut.drop(columns=['MRA_previous'])
    y = akut['MRA_previous']
    
    smote = SMOTE()
    x_smote , y_smote = smote.fit_resample(x, y)
    akut_smote = pd.concat([x_smote , y_smote], axis=1)

    print("SMOTE")
    print('Original akutset shape:', Counter(y))
    print('Resampple akutset shape:', Counter(y_smote))
    
    return akut_smote

###########################################################

def NearMiss_Under_Sampler(akut):
    
    x = akut.drop(columns=['MRA_previous'])
    y = akut['MRA_previous']
    
    nm = NearMiss()
    x_nm, y_nm = nm.fit_resample(x, y)
    akut_nm = pd.concat([x_nm, y_nm], axis=1)
    
    print("NMUS")
    print('Original akutset shape:', Counter(y))
    print('Resample akutset shape:', Counter(y_nm))
    
    return akut_nm
```

```python
akut_RUS = Random_Under_Sampler(akut)
```
