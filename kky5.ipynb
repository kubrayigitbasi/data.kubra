```python
# Kütüphanelerin Çağırılması
import numpy as np
import pandas as pd

import seaborn as sns
sns.set()

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Veriyi okumadan önce Sütun isimlerini içerisinde barındıran bir liste oluşturduk, kendi istediğimiz gibi isimlendirdik.
features = ['ID', 'Center_cat', 'Age', 'Gender', 'Weight', 'Height', 'BMI', 'NYHA',
       'FE', 'VO2_peak', 'VO2_peak_Kg', 'VO2_peak_perc', 'VE_VCO2_S', 'MDRD',
       'Crea', 'Na', 'K', 'Hb', 'Ac_Urico', 'FU_total', 'Exitus', 'causa',
       'ACE_I', 'AT1_I', 'BB', 'Tipo BB', 'Dose BB', 'Dose_equiv', 'Diuretici',
       'Diur_tipo', 'Dose diur', 'Diur2_tipo', 'Dose diur1', 'Statin',
       'Allopurinolo', 'Risparmiatore_K', 'Antiaggreganti', 'TAO', 'Digitale']


# Veriyi okurken names= features diyerek kendi veridiğimiz isimlere göre okuması için komut verdik.
KKY = pd.read_csv("https://raw.githubusercontent.com/kubrayigitbasi/data_kubra/main/KKY.csv" , names= features , header=0)
KKY.sample(10)
```
